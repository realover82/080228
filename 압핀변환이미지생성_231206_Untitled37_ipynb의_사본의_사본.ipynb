```python
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt

def create_grid_tilted_image(original_image, rectangles, grid_size=16, num_tilted=10, tilt_angle=45):
    image = original_image.copy()
    random_rect = random.choice(rectangles)
    print(random_rect)
    x, y, w, h = random_rect

    # Calculate step size for grid
    step_x = w // grid_size
    step_y = h // grid_size

    # Get all grid boxes
    grid_boxes = [(x + i * step_x, y + j * step_y, step_x, step_y) for i in range(grid_size) for j in range(grid_size)]

    # Randomly select grid boxes to tilt
    selected_boxes = random.sample(grid_boxes, num_tilted)

    # Perform tilt on selected grid boxes
    M = cv2.getRotationMatrix2D((step_x // 2, step_y // 2), tilt_angle, 1)
    for box in selected_boxes:
        bx, by, bw, bh = box
        roi = image[by:by+bh, bx:bx+bw]
        tilted_roi = cv2.warpAffine(roi, M, (bw, bh), borderMode=cv2.BORDER_REPLICATE)
        image[by:by+bh, bx:bx+bw] = tilted_roi

    return image

# Load the original image (you need to provide the path to your image)
# original_image = cv2.imread('/content/drive/MyDrive/deeplearning/remocon/original/5.jpg')
original_image = cv2.imread('/content/drive/MyDrive/deeplearning/remocon/original03/Snapshot_2023-11-13_15.25.51.jpg')
# Define the rectangles (you need to adjust this based on your image)
# rectangles = [
#     # (28, 28, 285, 281) #1번
#     # (57, 394, 285, 281) #2번
#     (828, 394, 285, 281)  #3번
#     # (57, 1804, 285, 281), #4
#     # (456, 1916, 285, 281), #5
#     # (828, 1804, 285, 281), #6
#     # (28, 2706, 285, 281), #7
#     # (428, 2515, 285, 281), #8
#     # (828, 2706, 285, 281), #9
#     # (428, 2881, 285, 281) # 10
# ]
rectangles = [

    # (328, 328, 200, 200) #1번
    # (257, 644, 285, 281) #2번
    # (1028, 644, 285, 281)  #3번
    # (307, 2004, 285, 281) #4
    # (656, 2116, 285, 281) #5
    # (1028, 2004, 285, 281) #6
    # (278, 2906, 285, 281) #7
    # (678, 2715, 285, 281) #8
    # (1078, 2906, 285, 281) #9
    # (678, 3081, 285, 281) # 10

    # (1628, 328, 200, 200) #11번
    # (1557, 644, 285, 281) #12번
    # (2328, 644, 285, 281)  #13번
    # (1607, 2004, 285, 281) #14
    # (1956, 2116, 285, 281) #15
    # (2328, 2004, 285, 281) #16
    # (1578, 2906, 285, 281) #17
    # (1978, 2715, 285, 281) #18
    # (2378, 2876, 285, 281) #19
    # (1978, 3081, 285, 281) # 20

    # (2928, 328, 200, 200) #21번
    # (2857, 614, 285, 281) #22번
    # (3628, 614, 285, 281)  #23번
    # (2907, 2004, 285, 281) #24
    # (3256, 2116, 285, 281) #25
    # (3628, 2004, 285, 281) #26
    # (2878, 2906, 285, 281) #27
    # (3278, 2715, 285, 281) #28
    # (3678, 2876, 285, 281) #29
    (3278, 3081, 285, 281) # 30

    # (4228, 278, 200, 200), #31번
    # (4157, 594, 285, 281), #32번
    # (4928, 594, 285, 281),  #33번
    # (4207, 1954, 285, 281), #34
    # (4556, 2066, 285, 281), #35
    # (4928, 1954, 285, 281), #36
    # (4178, 2856, 285, 281), #37
    # (4578, 2665, 285, 281), #38
    # (4978, 2856, 285, 281), #39
    # (4578, 3051, 285, 281) # 40


]



# Create and save 30 tilted images
for i in range(30):
    # Create tilted image based on the grid
    grid_tilted_image = create_grid_tilted_image(original_image, rectangles)

    # Display the grid tilted image using matplotlib (optional)
    # plt.figure(figsize=(10, 10))
    # plt.imshow(cv2.cvtColor(grid_tilted_image, cv2.COLOR_BGR2RGB))
    # plt.axis('off')
    # plt.show()

    # Save the image with a unique filename
    save_path = f"/content/drive/MyDrive/deeplearning/remocon/test03-02-3/tilled_image_button_30_{i + 1}.jpg"
    cv2.imwrite(save_path, grid_tilted_image)

print(original_image.shape)
```
