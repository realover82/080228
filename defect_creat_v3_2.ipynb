```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import random

# 원본 이미지를 읽어옵니다.
original_image = cv2.imread('/content/drive/MyDrive/deeplearning/remocon/original/defect_device-2-1.jpg')

if original_image is None:
    raise ValueError(f"Image not found at {original_image}")

rectangles = [
    (10, 70, 50, 50),
    (5, 5, 50, 50),
    (145, 70, 50, 50),
    (10, 320, 50, 50),
    (80, 340, 50, 50),
    (145, 320, 50, 50),
    (10, 480, 50, 50),
    (80, 455, 50, 50),
    (80, 520, 50, 50),
    (145, 480, 50, 50)
]

# 10장의 이미지를 생성합니다.
for i in range(10):
    # 원본 이미지의 복사본을 생성합니다.
    image = original_image.copy()

    # 경계상자 중에서 랜덤하게 하나를 선택합니다.
    random_rect = random.choice(rectangles)
    x, y, w, h = random_rect

    # 선택된 random_rect 값을 출력합니다.
    print(random_rect)

    roi = image[y:y+h, x:x+w]

    # 선택된 경계상자 안에 스크래치를 추가합니다.
    num_scratches = random.randint(1, 5)
    for _ in range(num_scratches):
        start_x = random.randint(0, w)
        start_y = random.randint(0, h)
        end_x = start_x + random.randint(-20, 20)
        end_y = start_y + random.randint(-20, 20)
        cv2.line(roi, (start_x, start_y), (end_x, end_y), (78, 84, 65), 2)

    # 결과 이미지를 표시합니다.
    cv2_imshow(image)

    # random_rect 값을 파일명으로 사용하여 결과 이미지를 저장합니다.
    save_path = f"/content/drive/MyDrive/deeplearning/remocon/test05/scratched_image_{random_rect}_{i}.jpg"
    cv2.imwrite(save_path, image)
```
